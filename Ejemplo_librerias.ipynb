```python
from scipy.fftpack import fft
import matplotlib.pyplot as plt
import numpy as np
```

```python
#numer of samplepoints
n = 1000

#sample spacing
dt = 1.0/1000


t = np.linspace(0.0,n*dt,n)
y = np.sin(2.0*np.pi*50*t) + np.sin(2.0*np.pi*100*t) + 4*np.sin(2.0*np.pi*20*t)

ft_y = fft(y)
```

```python
plt.plot(t[0:100],y[0:100])
plt.grid()
plt.show()
```

```python
plt.plot(t[0:150],ft_y[0:150])
plt.grid()
plt.show()
```
