```python
from scipy import signal
import matplotlib.pyplot as plt
```

```python
Rp = 1000
Rs = 1000
Cp = 1*(10**-6)
Rtot = 1/(1/Rs+1/Rp)
```

```python
sys = signal.TransferFunction([Rp/(Rs+Rp)],[Cp*Rtot,1])
w, mag, phase = signal.bode(sys)
```

```python
sys2 = signal.TransferFunction([Rp/(Rs+Rp)],[10*Cp*Rtot,1])
w2, mag2, phase2 = signal.bode(sys2)
```

```python
sys3 = signal.TransferFunction([Rp/(Rs+Rp)],[100*Cp*Rtot,1])
w3, mag3, phase3 = signal.bode(sys3)
```

```python
plt.figure()
plt.grid()
plt.semilogx(w,mag)
plt.semilogx(w2,mag2)
plt.semilogx(w3,mag3)
```

```python
plt.figure()
plt.grid()
plt.semilogx(w,phase)
plt.semilogx(w2,phase2)
plt.semilogx(w3,phase3)
```
